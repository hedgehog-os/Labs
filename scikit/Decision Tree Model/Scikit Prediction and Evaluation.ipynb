```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
```

## Loading data

```python
df = pd.read_csv('vehicles.csv')
df
```

## Cleaning

```python
df['Income'] = df['Income'].fillna(0.0)
df.head(8)
```

## Encoding

```python
df['Gender'].unique()
```

```python
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
df.head(10)
```

## Model

```python
X = df.drop(columns='Favorite Transport')
X.head(3)
```

```python
y = df['Favorite Transport']
y.head()
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
model
```

## Prediction

```python
test_df = pd.DataFrame({
    'Age': [12, 30, 75],
    'Gender':[0, 1, 0],
    'Income':[0.0, 35000, 300]
})
test_df
```

```python
model.predict(test_df)
```

## Exporting to the DOT file

```python
tree.export_graphviz(model, out_file='decision_tree_model.dot', feature_names=['Age', 'Gender', 'Income'], filled=True, class_names=sorted(y.unique()))
```

## Evaluation

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_test
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
model_accuracy = accuracy_score(y_test, predictions)
model_accuracy
```
